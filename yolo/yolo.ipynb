```python
import torch.nn as nn
import torchvision
import torch
from ultralytics import YOLO
```

```python
# Load YOLO
model = YOLO("../trained/best_yolo.pt")
```

```python
# Finetune it
results = model.train(
    data="yolo_info.yaml",
    epochs=50,
    imgsz=320,
    batch=2,
    device="cuda",
    hsv_h=0.05,
    hsv_s=0.45,
    hsv_v=0.25,
    degrees=5.0,
    translate=0.1,
    scale=0.3,
    flipud=0.25,
    fliplr=0.25,
)
```
